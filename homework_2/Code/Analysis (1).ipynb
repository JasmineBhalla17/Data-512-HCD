For our analysis, we will be producing 6 tables

```python
import pandas as pd
import numpy as np
```

Our entire analysis is on per-capita basis and for region and country. So, we create two dataframes for the same that can be used to produce the desired results.

```python


df = pd.read_csv("../Results/wp_politicians_by_country.csv")
```

```python
df.head()
```

```python
df.drop(columns=['Unnamed: 0'], inplace= True)
```

```python
df
```

We have to do our calculations in per-capita: per-capita is for example, total articles by a country divided by its total population.

```python
article_percapita_country = (df.groupby('country')['article_title'].count() / (df.groupby('country')['population'].mean()* 1000000)).to_frame(name = 'per_capita_article')


```

```python
article_percapita_country[:10]
```

```python
article_percapita_region = (df.groupby('region')['article_title'].count() / (df.groupby('region')['region_population (millions)'].mean()*1000000)).to_frame(name = 'per_capita_region').reset_index()
                           
                           
                    
```

```python
article_percapita_region[:10]
```

```python
# Now we do the same for article quality -- making dfs for high-quality-articles-per-population (per country and per region)
```

```python
high_quality_articles = df[df['article_quality'].isin(['FA', 'GA'])]
```

```python
high_quality_articles[:10]
```

```python
high_quality_percapita_country = (high_quality_articles.groupby('country')['article_title'].count() / (high_quality_articles.groupby('country')['population'].mean()*1000000)).to_frame(name = 'per_capita_article').reset_index()
                                    
                                  
```

```python
high_quality_percapita_country[:10]
```

```python
high_quality_percapita_region = (high_quality_articles.groupby('region')['article_title'].count() / (high_quality_articles.groupby('region')['region_population (millions)'].mean() * 1000000)).to_frame(name = 'per_capita_region').reset_index()


```

```python
high_quality_percapita_region[:10]
```

**Table 1: Top 10 countries by coverage i.e top 10 countries with the highest number of articles per-capita**

```python
article_percapita_country.sort_values(by = 'per_capita_article', ascending= False).replace([np.inf, -np.inf], np.nan).dropna().head(10)
```

**Bottom 10 countries by coverage**

The 10 countries with the lowest total articles per capita(in ascending order)

```python
article_percapita_country.sort_values(by= 'per_capita_article').head(10)
```

**Top 10 countries by High Quality**

The top 10 countries with the highest high quality articles per capita (in descending order).

```python
high_quality_percapita_country.sort_values(by = 'per_capita_article', ascending= False).head(10)
```

**Bottom 10 countries by High quality**

```python
high_quality_percapita_country.sort_values(by = 'per_capita_article').head(10)
```

**Geographic Regions by total coverage**

```python
article_percapita_region.sort_values(by = 'per_capita_region', ascending=False)
```

**Geogrpahic Regions by High Quality Coverage**

```python
# Rank ordered list of geographic regions(in descending order) by high quality articles per capita

high_quality_percapita_region.sort_values(by = 'per_capita_region', ascending = False)
```
